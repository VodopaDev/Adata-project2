```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pyarrow.parquet as pq
%matplotlib inline
```

```python
path = 'resources/extended_dataset.parquet'
df = pd.read_parquet(path)
df.head(2)
```

```python
community_areas = pd.read_csv('resources/community_areas_infos.csv', delimiter=',').set_index('COMMUNITY AREA NAME')
community_areas.index = community_areas.index.str.upper()
community_areas.head()
```

```python
mean_nb_violations_by_community = df[['Community', 'Nb Violations', 'Nb Violations of high severity', 'Nb Violations of medium severity', 'Nb Violations of low severity']].groupby('Community').mean()
mean_nb_violations_by_community.head()
```

```python
df_joinded = mean_nb_violations_by_community.join(community_areas)
df_joinded.head()
```

```python
for column_name in community_areas.columns:
    display(df_joinded[['Nb Violations', column_name]].corr())
```

Sadly we have absolutly no correlation :(
